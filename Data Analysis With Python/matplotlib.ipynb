```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
sales_data_df = pd.read_csv("E:\\UDemy Data Analytics\\Complete-Python-Bootcamp-main\\Complete-Python-Bootcamp-main\\10-Data Analysis With Python\\sales_data.csv")

# Convert 'Date' to datetime if not already
sales_data_df['Date'] = pd.to_datetime(sales_data_df['Date'])

# Create 'Year' and 'Month' columns
sales_data_df['Year'] = sales_data_df['Date'].dt.year
sales_data_df['Month'] = sales_data_df['Date'].dt.month

# Group by Year and Month
sales_trend = sales_data_df.groupby(['Year', 'Month'])['Total Revenue'].sum().reset_index()

# Create a 'YearMonth' column for x-axis labels
sales_trend['YearMonth'] = sales_trend['Year'].astype(str) + '-' + sales_trend['Month'].astype(str).str.zfill(2)

# Plot the sales trend
plt.figure(figsize=(10,6))
plt.plot(sales_trend['YearMonth'], sales_trend['Total Revenue'], marker='o')
plt.xlabel('Year-Month')
plt.ylabel('Total Revenue')
plt.title('Sales Trend by Month')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
sales_trend=sales_data_df.groupby('Date')['Total Revenue'].sum().reset_index()
print(sales_trend)
```

```python
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.title('Sales by date')
plt.show()
```
