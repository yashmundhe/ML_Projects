```python
#Create a set with the first 10 positive integers. Print the set.
s = set(range(1,11))
print(s)
```

```python
#Add the number 11 to the set created in Assignment 1. Then remove the number 1 from the set. Print the modified set.
s.add(11)
s.remove(1)
print(s)
```

```python
#Create two sets: one with the first 5 positive integers and another with the first 5 even integers. 
#Perform and print the results of union, intersection, difference, and symmetric difference operations on these sets.
a = set(range(1,6))
b = set(range(2,12,2))
print(a)
print(b)
union = a.union(b)
print(union)
intersection = a.intersection(b)
print(intersection)
difference = a.difference(b)
print(difference)
sym_diff = a.symmetric_difference(b)
print(sym_diff)
```

```python
#Create a new set containing only the even numbers from the set created in Assignment 1 using a set comprehension. 
# Print the new set.
even_num = {x for x in s if x%2==0}
print(even_num)
```

```python
#Create a set with duplicate elements and remove the duplicates using set methods. Print the modified set.
p = {1,22,33,33,23,22}
unique = set(p)
print(p)
```

```python
#Create two sets: one with the first 5 positive integers and another with the first 3 positive integers. 
# Check if the second set is a subset of the first set and if the first set is a superset of the second set. 
# Print the results.
f= {1,2,3,4,5}
g={1,2,3}

if g.issubset(f):
    print('True')
else:
    print('False')

if f.issuperset(g):
    print('True')
else:
    print('False')
```

```python
#Create a frozenset with the first 5 positive integers. Print the frozenset.
h = frozenset(range(1,6))
print(h)
```

```python
#Create a set with the first 5 positive integers. 
#Convert it to a list, append the number 6, and convert it back to a set. 
#Print the resulting set.

j =set(range(1,6))
h = list(j)
h.append(6)
j=set(h)
print(j)
```

```python
#Create a dictionary with set keys and integer values.
#Print the dictionary.
k = {
    frozenset({1,2}) : {3},
    frozenset({4,5}) : {6},
}
print(k)
print(type(k))
```

```python
#Create a set and iterate over the elements, printing each element.
l = set(range(1,21))
# print(l)

for i in l:
    print(i, end=" ")
```

```python
#Create a set and remove elements from it until it is empty. Print the set after each removal.
m = set(range(0,16))
while m:
    m.pop()
    print(m)
```
