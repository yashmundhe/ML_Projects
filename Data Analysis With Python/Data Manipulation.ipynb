```python
df_filled=dataset.fillna(0)
```

```python
dataset['Sales_fillNA']=dataset['Sales'].fillna(dataset['Sales'].mean())
dataset
```

```python
dataset.dtypes
```

```python
dataset= dataset.rename(columns={'Date':'Sales Date'})
dataset.head()
```

```python
dataset['Value_new']=dataset['Value'].fillna(dataset['Value'].mean()).astype(int)
dataset
```

```python
dataset['New Value']=dataset['Value'].apply(lambda x:x**2)
dataset.head()
```

```python
grouped_mean=dataset.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
grouped_sum=dataset.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```

```python
grouped_mean=dataset.groupby(['Product','Region'])['Value'].mean()
grouped_mean
```

```python
grouped_agg=dataset.groupby('Region')['Value'].agg(['mean','sum','count'])
grouped_agg
```

```python
df1 = pd.DataFrame({'Key':['A','B','C'], 'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['C', 'E', 'F'], 'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='Key',how='inner')
```

```python
pd.merge(df1,df2,on='Key',how='outer')
```

```python
pd.merge(df1,df2,on='Key',how='left')
```

```python
pd.merge(df1,df2,on='Key',how='right')
```
