```python
#Reading large file

def read_large_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line
```

```python
for line in read_large_file(file_path='large_file.txt'):
    print(line.strip())
```

```python
def main_welcome(func):

    def sub_welcome_method():
        print('Welcome to the advanced python course')
        func()
        print('Hope you study well')
    return sub_welcome_method()
```

```python
def course_intro():
    print('this is an ML course')
course_intro()
```

```python
def my_decorator(func):
    def wrapper():
        print('H, Hola')
        func()
        print('Thanks')
    return wrapper()
```

```python
@my_decorator
def greet():
    print('heyyyy')
```

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print('Hello')
```

```python
say_hello()
```
