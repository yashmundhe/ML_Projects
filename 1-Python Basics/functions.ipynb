```python
def print_details(*args,**kwargs):
    for i in args:
        print(f"Positional argument: {i}")
    for key, value in kwargs.items():
        print(f"Keyword argument: {key}: {value}")
```

```python
print_details(1,2,3,"Krish",name="Abraham", age=23)
```

```python
def temp_converter(num, unit):
    if unit == 'f':
        return (num-32)*0.56
    elif unit == 'c':
        return (num*1.8)+32
    else:
        return "Invalid"
```

```python
temp_converter(12, 'f')
```

```python
def lift_capacity(n, capacity):
    if  n%capacity == 0:
        return n//capacity
    elif n%capacity != 0:
        return n//capacity+1
    else:
        return 'None'
```

```python
lift_capacity(12, 11)
```
