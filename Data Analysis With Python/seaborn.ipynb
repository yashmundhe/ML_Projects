```python
import seaborn as sns
```

```python
#Basic plotting with seaborn
tips=sns.load_dataset('tips')
tips
```

```python
#create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip',data=tips)
plt.title("Scatter Plot of a Total Bill VS Tip")
plt.show()
```

```python
#lineplot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Line plot of total bill size')
plt.show()
```

```python
#categorical plots
#bar plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot by total bill by day')
plt.show
```

```python
#boxplot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Box plot of Total bill by Day')
plt.show()
```

```python
#violinplot
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('Box plot of Total bill by Day')
plt.show()
```

```python
### Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
plt.show()
```

```python
#kde Plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
#Pairplot
sns.pairplot(tips)
plt.show()
```

```python
#Heatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd
sales_df=pd.read_csv('E:\\UDemy Data Analytics\\Complete-Python-Bootcamp-main\\Complete-Python-Bootcamp-main\\10-Data Analysis With Python\\sales_data.csv')
sales_df.head(5)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y='Total Revenue',data=sales_df,estimator=sum)
plt.title('Total Sales by Product')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.barplot(x='Region', y='Total Revenue',data=sales_df,estimator=sum)
plt.title('Sales by region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Group data by region and sum total revenue
region_sales = sales_df.groupby('Region')['Total Revenue'].sum()

# Create a pie chart
plt.figure(figsize=(5, 5))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales by Region')
plt.show()
```
