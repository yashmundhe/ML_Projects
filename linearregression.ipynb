```python
## Simple Calculator

num1 = float(input("Enter the first number "))
operation = input("Enter the operation here ")
num2 = float(input("Enter the second number "))


if operation == "+":
    print("The addition of the two numbers is:", num1+num2)
elif operation == "-":
    print("The subtraction of the two numbers is: ", num1-num2)
elif operation == "*":
    print("The multiplication of the two numbers is: ", num1*num2)
elif operation == "/":
    if num2 ==0:
        print("Error, Division by zero")
    else:
        print("The division of the two numbers is: ", num1/num2)
elif operation == "//":
    print("The floor division of the two numbers is: ", num1//num2)
elif operation == "%":
    print("The remainder agter multpliaction is: ", num1%num2)
elif operation == "**":
    print("The exponential of two numbers is:", num1**num2)
else:
    print("Invalid Operation")
```
